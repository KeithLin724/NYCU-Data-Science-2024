```python
import pandas as pd 
import shutil
import os 
```

```python
all_data_df = pd.read_csv("./tran_data_all.csv")
all_data_df
```

```python
HOT_FOLDER , BOO_FOLDER = "./data/HOT" , "./data/BOO"

if not os.path.exists(HOT_FOLDER):
    os.makedirs(HOT_FOLDER)
    
if not os.path.exists(BOO_FOLDER):
    os.makedirs(BOO_FOLDER)
```

```python
def move_file_func(data_dict):
    if data_dict["state_code"] == 200 and data_dict["reason_phase"] =="OK" and data_dict["image_state"] and data_dict["file_path"] != "":
        # to folder
        to_folder = HOT_FOLDER if data_dict["is_hot"] == 1 else BOO_FOLDER
        
        # is move it file
        if not os.path.exists(data_dict["file_path"]):
            _ , file_name = os.path.split(data_dict["file_path"])
            
            check_path = os.path.join(to_folder , file_name)
            
            # have this file 
            if os.path.exists(check_path):
                return {"image_link" : data_dict["image_link"] , "new_path" : check_path}
            
            
        try:
            new_path = shutil.move(data_dict["file_path"] , to_folder)  
        except Exception as e:
            # same file
            print(e)
            _ , file_name = os.path.split(data_dict["file_path"])
            
            new_path = os.path.join(to_folder , file_name)
            return {"image_link" : data_dict["image_link"] , "new_path" : new_path}
        
        # have new path
        return {"image_link" : data_dict["image_link"] , "new_path" : new_path}
    
    
    # no new_path
    return {"image_link" : data_dict["image_link"] , "new_path" : ""}



```

```python
all_data_list = all_data_df.to_dict("records")
all_data_list
```

```python
new_path_list = [move_file_func(item) for item in all_data_list]
new_path_list
```

```python
new_path_df = pd.DataFrame(new_path_list)
new_path_df
```

```python
new_path_df.to_csv("new_path.csv")
```

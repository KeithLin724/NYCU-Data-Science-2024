```python
import pandas as pd 
import os 
import numpy as np
```

```python
# all_table = pd.read_csv("./table/train_all_data_new.csv")
all_table = pd.read_csv("./table/image_table_new_v2.csv")
all_table
```

```python
all_table.loc[:,"is_hot"] = all_table.apply(lambda row : row["hotNumber"] > 35 , axis=1)
all_table
```

```python
image_table = all_table[all_table["image_status"] == True]
image_table
```

```python
all_table.value_counts("hotNumber").sort_index().plot.bar()
```

```python
image_table.value_counts("is_hot").sort_index().plot.pie()
```

```python
hot_image_table = image_table[image_table["is_hot"]]
hot_image_table
```

```python
hot_image_series = hot_image_table.value_counts("hotNumber").sort_index()
hot_image_list = list(hot_image_series.items())

hot_range_dict = {f"{hot_image_list[0][0]}_{hot_image_list[-1][0]}":sum(hot_image_series)}
hot_range_dict
```

```python
boo_image_table = image_table[image_table["is_hot"] == False]
boo_image_table
```

```python
boo_number_series = boo_image_table.value_counts("hotNumber").sort_index()
boo_number_series
```

```python
boo_number_series.plot.bar()
```

```python
range_dict  = {}
start = None
range_contain_size = 11000 , 12000
contain = 0 
for key , value  in boo_number_series.items():
    if start is None:
        start = key 
        
    contain += value
    if  range_contain_size[0] <= contain <= range_contain_size[1] :
        # too large 
        range_dict |= {f"{start}_{key}":contain}
        start = None
        contain = 0
    elif contain > range_contain_size[1]:
        range_dict |= {f"{start}_{key}":contain}
        start = None
        contain = 0
        
range_dict |= {f"{start}_{key}":contain}
    
```

```python
range_dict
```

```python
range_dict |= hot_range_dict
range_dict
```

```python
len(image_table) == sum(range_dict.values())
```

```python
range_key = [list(map(int , item.split("_"))) for item in range_dict.keys()]
range_key
```

```python
def range_type(row:dict)->str:
    for key in range_key:
        if  key[0] <= row["hotNumber"] <= key[1]:
            return f"{key[0]}_{key[1]}"
    
    return ""
```

```python
image_table.loc[:,"in_range"] = image_table.apply(range_type , axis=1)
image_table
```

```python
image_table
```

```python
for folder in range_dict.keys():
    folder_path = os.path.join("./database" , folder)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(folder_path)
```

```python
image_table = image_table.reset_index()
image_table
```

```python
image_table.loc[:,"to_folder"] = image_table.apply(lambda row: os.path.join(row["in_range"] , row["path"]) , axis=1)
image_table
```

```python
import shutil
```

```python
def move_file_by_row(row:dict):
    try:
        src_path , dest_path = row["path"] , row["to_folder"]
        full_path = lambda path : os.path.join("./database" , path)
        
        shutil.move(full_path(src_path) , full_path(dest_path))
    except Exception as e:
        return str(e)
    
    return None
    
```

```python
image_table.loc[:,"is_success"] = image_table.apply(move_file_by_row , axis=1)
image_table
```

```python
image_table = image_table.drop(columns=["to_folder" , "is_success"])
image_table
```

```python
image_table.to_csv("./table/dataset_table.csv")
```

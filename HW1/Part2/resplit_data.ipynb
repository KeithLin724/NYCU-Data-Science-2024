```python
all_table = pd.read_csv("./table/train_all_data.csv")
all_table
```

```python
hot_table = all_table[(all_table["hotNumber"] > 35) & (pd.notna(all_table["new_path"]))]
hot_table
```

```python
boo_table = all_table[(all_table["hotNumber"] <35) & (pd.notna(all_table["new_path"]))]
boo_table
```

```python
boo_hot_number = boo_table["hotNumber"]
boo_hot_number
```

```python
hot_numbers_count = boo_hot_number.value_counts().sort_index()
```

```python
range_dict  = {}
start = None
range_contain_size = 11000 , 12000
contain = 0 
for key , value  in hot_numbers_count.items():
    if start is None:
        start = key 
        
    contain += value
    if  range_contain_size[0] <= contain <= range_dict[1] :
        # too large 
        pass 
        
    
```

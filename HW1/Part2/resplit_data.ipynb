```python
import pandas as pd 
import os 
import numpy as np
```

```python
all_table = pd.read_csv("./table/train_all_data_new.csv")
all_table
```

```python
all_table.value_counts("hotNumber").sort_index().plot.bar()
```

```python
image_table = all_table[all_table["image_state"] == True]
image_table
```

```python
hot_image_table = image_table[image_table["is_hot"] == 1]
hot_image_table
```

```python
hot_image_series = hot_image_table.value_counts("hotNumber").sort_index()
hot_image_list = list(hot_image_series.items())

hot_range_dict = {f"{hot_image_list[0][0]}/{hot_image_list[-1][0]}":sum(hot_image_series)}
hot_range_dict
```

```python
boo_image_table = image_table[image_table["is_hot"] == 0]
boo_image_table
```

```python
boo_number_series = boo_image_table.value_counts("hotNumber").sort_index()
boo_number_series
```

```python
boo_number_series.plot.bar()
```

```python
range_dict  = {}
start = None
range_contain_size = 11000 , 12000
contain = 0 
for key , value  in boo_number_series.items():
    print(key, value)
    if start is None:
        start = key 
        
    contain += value
    if  range_contain_size[0] <= contain <= range_contain_size[1] :
        # too large 
        range_dict |= {f"{start}/{key}":contain}
        start = key
        contain = 0
    elif contain > range_contain_size[1]:
        range_dict |= {f"{start}/{key}":contain}
        start = key
        contain = 0
        
range_dict |= {f"{start}/{key}":contain}
    
```

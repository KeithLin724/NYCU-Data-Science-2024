```python
from CrawlerHW import CrawlerHW
from CrawlerData import CrawlerData
```

```python
async def check_image(record:dict , client:httpx.AsyncClient):
    check_folder = "BOO" if (record["is_hot"] == 0)  else "HOT"

    if check_folder not in record["new_path"]:
        print(f"something wrong {record['title']} , {record['url']}")
        return record | {"error": "not in current folder"}
    
    # normal
    if os.path.exists(record["new_path"]):
        return None
    
    # no this image 
    await asyncio.sleep(CrawlerHW.get_random_wait_time())
    
    try:
        image_content = await client.get(record['url'], headers=CrawlerHW.get_random_user_agent(), timeout=60)
    except Exception as e:
        print(e)
        return record | {"error" : str(e)}
    
    # can not get image
    if image_content.status_code != 200:
        record["state_code"] = image_content.status_code
        return record | {"error" : f"status_code: { image_content.status_code}"}
    
    # save image
    image = CrawlerData.bytes_to_image(image_content.content)
    try:
        cv2.imwrite(record["new_path"] , image)
    except Exception as e:
        print(e)
        return record | {"error" : f"write image error :{str(e)}"}
        
    
    # is get image
    return None
```

```python
async def add_image_from_table(table:pd.DataFrame):
    records = table.to_dict("records")
    
    async with httpx.AsyncClient() as client:
        tasks = [check_image(record , client) for record in records]

        error_task = await CrawlerHW.gather(*tasks)
    
    return [item for item in error_task if item is not None]
```

```python
hot_error_records = await add_image_from_table(hot_table)
```

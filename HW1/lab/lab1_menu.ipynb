```python
from bs4 import BeautifulSoup
import httpx

URL = "https://www.ptt.cc/bbs/Beauty/index.html"
header = {"cookie": "over18=1"}
```

```python


result = httpx.get(URL , headers=header)
```

```python
soup = BeautifulSoup(result.text , "html.parser")
soup
```

```python
# need next and old html

action_bar_button = soup.find_all("div", class_="btn-group btn-group-paging")[0]
action_bar_button = action_bar_button.find_all("a")
action_bar_button
```

```python
# action_bar_button = {}

location = ["Old" , "Prev" , "Next" , "New"]
button_link = [item.get("href" , "") for item in action_bar_button]
button_link

button_dict = dict(zip(location , button_link))

button_dict
```

```python

recommend_list = soup.find("div" , class_="r-list-container action-bar-margin bbs-screen")
recommend_list
```

```python
recommend = recommend_list.find_all("div" , class_="r-ent")
print(len(recommend))
recommend
```

```python

def change_to_dict(div_obj:BeautifulSoup):    
    hot_number = div_obj.find("span")
    
    hot_number = hot_number.string if hot_number is not None else 0
    
    title = div_obj.find("a")
    title_str , title_url = "" , ""
    
    if title is not None:
        title_str , title_url =   title.string , title.get("href")
        
    
    author = div_obj.find("div" , class_="author")
    author = str(author.string) if author is not None else ""
    
    date = div_obj.find("div" , class_="date")
    date = str(date.string) if date is not None else ""
    
    return {
        "Title":title_str,
        "Author" : author,
        "HotNumber" : hot_number ,
        "Date" : date,
        "URL":title_url,
    }

change_to_dict(recommend[0])
```

```python
recommend_result = [change_to_dict(div_obj=item) for   item in recommend]
recommend_result
```

```python
def filter_function(item_dict:dict):
    if "[公告]" in item_dict["Title"]:
        return False 
    
    if "Fw:[公告]" in item_dict["Title"]:
        return False 
    
    if item_dict["Title"] == "":
        return False 
    
    if item_dict["URL"] == "":
        return False 
    
    return True


recommend_after_filter = list(filter(filter_function , recommend_result))
recommend_after_filter
```

```python
recommend_page_dict = button_dict | {"Body" : recommend_after_filter}
recommend_page_dict
```
